```python
# Importing Important libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
original_data = pd.read_csv("diabetes_health_data.csv")
original_data.head(10)
```

Link to data [here](https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset?select=diabetes_binary_health_indicators_BRFSS2015.csv).

## Meta Data
- Diabetes_binary: 0 = no diabetes 1 = prediabetes 2 = diabetes
- HighBP: 0 = no high BP 1 = high BP
- HighChol: 0 = no high cholesterol 1 = high cholesterol ***Normal: Less than 200 mg/dL HIGH 200+***
- CholCheck: 0 = no cholesterol check in 5 years 1 = yes cholesterol check in 5 years. 
- BMI: Body Mass Index - ***an approximate measure of whether someone is over- or underweight, calculated by dividing their weight in kilograms by the square of their height in metres.***
- Smoker: Have you smoked at least 100 cigarettes in your entire life? [Note: 5 packs = 100 cigarettes] 0 = no 1 = yes
- Stroke: (Ever told) you had a stroke. 0 = no 1 = yes
- HeartDiseaseorAttack: coronary heart disease (CHD) or myocardial infarction (MI) 0 = no 1 = yes
- PhysActivity: physical activity in past 30 days - not including job 0 = no 1 = yes
- Fruits: Consume Fruit 1 or more times per day 0 = no 1 = yes
- Veggies: Consume Vegetables 1 or more times per day 0 = no 1 = yes
- HvyAlcoholConsump: (adult men >=14 drinks per week and adult women>=7 drinks per week) 0 = no 1 = yes
- AnyHealthcare: Have any kind of health care coverage, including health insurance, prepaid plans such as HMO, etc. 0 = no 1 = yes
- NoDocbcCost: Was there a time in the past 12 months when you needed to see a doctor but could not because of cost? 0 = no 1 = yes
- GenHlth: Would you say that in general your health is: scale 1-5 1 = excellent 2 = very good 3 = good 4 = fair 5 = poor
- MentHlth: days of poor mental health scale 1-30 days
- PhysHlth: physical illness or injury days in past 30 days scale 1-30
- DiffWalk: Do you have serious difficulty walking or climbing stairs? 0 = no 1 = yes
- Sex: 0 = female 1 = male
- Age: Individuals age
- Education: Education level
- Income: Individuals income.

## Beginning of EDA

```python
# Number of Positive VS Negative people.
print("The number of Diabetes positive and negative samples are: ")
negative_postive = original_data.Diabetes_binary.value_counts()
print(f"Diabetes Negative: {negative_postive[0]}\nDiabetes Positive: {negative_postive[1]}")
plt.bar(["Negative", "Positive"], [negative_postive[0], negative_postive[1]])
plt.title("Bar Plot of Diabetes Negative and Positive results", fontsize= 15)
plt.xlabel("Results")
plt.ylabel("Amount")
plt.show()
```

```python
plt.figure(figsize= (15, 10))
plt.title("Heat map showing relationship of data attributes", fontsize= 15)
sns.heatmap(original_data.corr(), cmap= "Blues")
plt.show()
```

```python
# Evaluating BMI in order to detect and drop outliers.

plt.figure(figsize=(18,7))
plt.subplot(1,2,1)
plt.boxplot(original_data["BMI"], vert= False)
plt.title("BMI Box-plot", fontsize=20)
plt.xlabel("BMI Score", fontsize= 15)

# Calculating lower and upper quartiles
lower_quartile_BMI = original_data.BMI.quantile(0.25)
upper_quartile_BMI = original_data.BMI.quantile(0.75)

# Bar plot showing average diatebes potitive and negative individuals BMI level
BMI_positive_negative = original_data.groupby("Diabetes_binary")["BMI"].mean()
plt.subplot(1,2,2)
plt.title("Average BMI of individuals", fontsize= 20)
plt.bar(["Negative", "Positive"], BMI_positive_negative.values)
plt.xlabel("Status", fontsize= 15)
plt.ylabel("BMI Score", fontsize= 15)
plt.ylim(0, 40)
plt.show()

print("BMI lower quantile score is %d.\nBMI upper quartile score is %d." %(lower_quartile_BMI, upper_quartile_BMI))
```

```python
# Removal of rows with false BMI scores.
data_2 = original_data[(original_data.BMI < upper_quartile_BMI) & (original_data.BMI > lower_quartile_BMI)]

print("New dataset Positive and Negative data count")
negative_postive_2 = data_2.Diabetes_binary.value_counts()
print(f"Diabetes Negative: {negative_postive_2[0]}\nDiabetes Positive: {negative_postive_2[1]}")
data_2.head()
```

```python
print()
data_2.describe().T
```

```python
# Number of Positive VS Negative people.
print("The number of Diabetes positive and negative samples are: ")
negative_postive_2 = data_2.Diabetes_binary.value_counts()
print(f"Diabetes Negative: {negative_postive_2[0]}\nDiabetes Positive: {negative_postive_2[1]}")
plt.bar(["Negative", "Positive"], [negative_postive_2[0], negative_postive_2[1]])
plt.title("Bar Plot of Diabetes Negative and Positive results", fontsize= 15)
plt.xlabel("Results")
plt.ylabel("Amount")
plt.show()
```

```python
# Undersampling majority class
negative_sample = data_2[data_2["Diabetes_binary"]==0].sample(n = 13014) # Selecting random sample the size of minority class.
positive_sample = data_2[data_2["Diabetes_binary"]==1]


negative_sample.shape
print(f"The data shape for both negative and positive data is: \nNegative: {negative_sample.shape}\nPositive: {positive_sample.shape}")
```

```python
# Congatenation of both data samples
data_concat = pd.concat((negative_sample, positive_sample), axis=0, ignore_index= True)
data_concat.head()
```

```python
data_concat.describe().T
```

```python
# Number of Positive VS Negative people.
print("The number of Diabetes positive and negative samples are: ")
negative_postive_concat = data_concat.Diabetes_binary.value_counts()
print(f"Diabetes Negative: {negative_postive_concat[0]}\nDiabetes Positive: {negative_postive_concat[1]}")
print("Data samples are now of equal amounts.... Time for further analysis")
plt.bar(["Negative", "Positive"], [negative_postive_concat[0], negative_postive_concat[1]])
plt.title("Bar Plot of Diabetes Negative and Positive results", fontsize= 15)
plt.xlabel("Results")
plt.ylabel("Amount")
plt.show()
```

```python
plt.figure(figsize= (15, 10))
plt.title("Heat map showing relationship of data attributes", fontsize= 15)
sns.heatmap(data_concat.corr(), cmap= "Blues")
plt.show()
```
